<h1>Quadratic Root Finding</h1>

```python
def roots(a,b,c):
    a,b,c=float(a),float(b),float(c)
    D=(b**2 -4*a*c)**0.5
    x_1=(-b+D)/2*a
    x_2 =(-b-D)/2*a
    print('x1: {0}'.format(x_1))
    print('x2: {0}'.format(x_2))
```

```python
roots(5,6,7)
```

<h1>Series</h1>

```python
from sympy import Symbol,pprint,init_printing
def print_series(n,x_value):
    init_printing(order="rev-lex")
    n=int(n)
    x=Symbol('x')
    series=x
    x_value=float(x_value)
    for i in range(2,n+1):
        series=series+(x**i)/i
    pprint(series)
    series_value = series.subs({x:x_value})
n=0
x_value=0
print_series(n,x_value)
    
```

<h1 style = "color: Blue;"> Linear Equations </h1>

```python
from sympy import solve,pprint,Symbol
def LinearSolve(exp1,exp2):
    result =solve((exp1,exp2),dict =True)
    pprint(result)
x=Symbol('x')
y=Symbol('y')
ex1=2*x+y*8+4
ex2=4*y+2*x+3
LinearSolve(ex1,ex2)
    
```

```python
from sympy.plotting import plot3d as plt
def plot_eq(ex1,ex2):
    plt(ex1,ex2)
plot_eq(ex1,ex2)
```

```python
from sympy import Symbol, solve, solve_univariate_inequality, sin
x = Symbol('x')
ineq_obj = sin(x) - 0.6 > 0
solve_univariate_inequality(ineq_obj, x, relational=False)
```
