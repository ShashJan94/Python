```python
from sklearn.linear_model import LinearRegression
observations =len(df)
linear_regression = LinearRegression(normalize=False,fit_intercept =True)
cubic = PolynomialFeatures(degree =3, interaction_only =False, include_bias=False)
quadratic = PolynomialFeatures(degree =2,interaction_only=False,include_bias =False)
linear_predictor =make_pipeline(linear_regression)
quardretic_predictor=make_pipeline(quadratic,linear_regression)
cubic_predictor = make_pipeline(cubic,linear_regression)
```

```python
import numpy as np
Predictor = 'carat'
x_range = [df['carat'].min(),df['carat'].max()] 
y_range = [df['price'].min(),df['price'].max()]
print(df['price'].max())
xr = df['carat'].values.reshape((observations,1))
xt = np.linspace(0,18500,53940).reshape((53940,1))
yr = df['price'].values.reshape((observations,1))
scatter = df.plot(kind='scatter', x='carat', y='price', xlim=x_range, ylim=y_range)
regr_line = scatter.plot(xt, cubic_predictor.fit(xr,yr).predict(xt), '-', color='red', linewidth=2)
```
